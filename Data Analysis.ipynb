```python
import matplotlib.pyplot as plt
import pandas as pd
fig, ax=plt.subplots()

count = df.recclass.value_counts(normalize=False)

threshold = 200
mask = count > threshold
tail_prob = count.loc[~mask].sum()
count = count.loc[mask]

count.plot(kind='bar',color='lightcoral',figsize=(20, 10))
plt.xticks(rotation=45)
plt.yticks(rotation=0)
#count.plot(figsize=(20, 10))
plt.xlabel("Meteorite Class",fontsize=20)
plt.ylabel('Class Count',fontsize=20)

plt.title('Meteorite Classification',fontsize=20)
ax.grid(which='major',linestyle='-',linewidth='0.2',color='darkgreen')
ax.tick_params(axis="x", labelsize=17)
ax.tick_params(axis="y", labelsize=17)

for i in ax.patches:
    plt.text(i.get_x()+0.05, i.get_y()+0.5,
             i.get_height(),
             fontsize = 14,  color ='black',verticalalignment='bottom')
   
plt.show()
```
